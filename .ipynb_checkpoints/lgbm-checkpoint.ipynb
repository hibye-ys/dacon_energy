```python
import lightgbm as lgb
import optuna
from tqdm import tqdm
import pickle

best_mape = float('inf') 
i = 0
n_trial = 3

# Turn off Optuna logging
optuna.logging.set_verbosity(optuna.logging.WARN)

def save_parameters(study, trial):
    with open(path + "lgb_parameters_log.txt", "a") as f:
        f.write(f"Trial {trial.number}, Params: {trial.params}, Value: {trial.value}, mape: {best_mape}\n")

def objective(trial):
    global i, best_mape
    
    params = {
        'objective': 'regression',
        'metric': 'rmse',
        'verbosity': -1,
        'boosting_type': 'dart',
        'max_depth': trial.suggest_int('max_depth', 1, 10),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.01, 1.0),
        'lambda_l2': trial.suggest_loguniform('lambda_l2', 0.1, 1.0),
        'lambda_l1': trial.suggest_loguniform('lambda_l1', 0.1, 1.0),
        'feature_fraction': trial.suggest_float('feature_fraction', 0.1, 1.0),
        'bagging_fraction': trial.suggest_float('bagging_fraction', 0.1, 1.0),
    }

    model = lgb.train(params, dtrain, valid_sets=[dval], early_stopping_rounds=100, verbose_eval=False)
    preds = model.predict(X_val, num_iteration=model.best_iteration)
    current_mape = mean_absolute_percentage_error(y_true=y_val, y_pred=preds)
    
    print("{} trial, MAPE: {}".format(i, current_mape))
    
    if current_mape < best_mape:
        best_mape = current_mape
        with open("lgb_best.pickle", "wb") as fout:
            pickle.dump(model, fout)
    else:
        pass
    
    i += 1
    
    return best_mape

progress_bar = tqdm(total=n_trial, desc="Optimizing", position=0)

study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=n_trial, callbacks=[lambda study, trial: progress_bar.update(1), save_parameters])

progress_bar.close()

print(study.best_trial)
print(study.best_value)
print("Best trial:")
trial = study.best_trial
print("  Value: ", trial.value)
for key, value in trial.params.items():
    print("    {}: {}".format(key, value))
```
